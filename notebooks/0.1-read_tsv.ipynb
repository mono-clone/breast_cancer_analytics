```python
import os
from pathlib import Path

import numpy as np
import pandas as pd

import config

#最大表示列数の指定（ここでは50列を指定）
pd.set_option('display.max_columns', 50)

def make_dir(dir_name: str):
    if not os.path.exists(dir_name):
        os.makedirs(dir_name)
```

## data_clinical_patient

```python
make_dir(config.INTERIM_BRCA_METABRIC_DIR+'/head')
```

```python

def create_head_csv(file_name: str, size: int=5):
    pd.read_table(config.RAW_BRCA_METABRIC_DIR + "/"+ file_name+".txt").head(size).to_csv(config.INTERIM_BRCA_METABRIC_DIR+'/head/'+ file_name+'_head.csv')
    
def show_head(file_name: str, size: int=5):
    display(pd.read_table(config.RAW_BRCA_METABRIC_DIR + "/"+ file_name+".txt").head(size))
    
def show_tail(file_name: str, size: int=5):
    display(pd.read_table(config.RAW_BRCA_METABRIC_DIR + "/"+ file_name+".txt").tail(size))
```

```python
p = Path(config.RAW_BRCA_METABRIC_DIR)

for file_name in sorted([str(file_name) for file_name in list(p.glob("**/data_*.txt"))]):
    file_name = file_name.split("/")[4].split(".")[0]
    print(file_name)
    create_head_csv(file_name, 4)
    show_head(file_name, 8)
    show_tail(file_name)
```
